```python
import math

x = []
y = []
for i in range(13):
    x.append(i+1)
    y.append(math.e**(i+1))
    print(str(i+1)+' --> '+str(y[i]))

plt.figure(figsize=(20,6))
    
plt.plot(x, y)
plt.title('e elevado a los 13 primeros números naturales', fontdict={'fontname': 'DejaVu Sans Mono', 'fontsize': 14})
plt.xlabel('Números naturales 1 a '+str(i+1), fontdict={'fontname': 'DejaVu Sans Mono', 'fontsize': 14})
#plt.xticks([1, 3, 5, 7, 9, 6, 7, 8, 9, 10, 11, 12, 13])
plt.ylabel('Resultado', fontdict={'fontname': 'DejaVu Sans Mono', 'fontsize': 14})

plt.show()
```

```python
gas = pd.read_csv("./datos/gas_prices.csv",sep=",")
gas.head(100)
```

```python
plt.figure(figsize=(20,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='S.Korea')
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')
plt.plot(gas.Year, gas.France, 'c.-', label='France')
plt.plot(gas.Year, gas.Italy, 'r.--', label='Italy')


# Another Way to plot many values!
# countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], marker='.')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```
